```python
from Geometry import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
cylinder = Cylinder([0, 0, 0], 3, 2, q=[0, np.pi / 4, 0])
points = cylinder.grid_points(50)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')

ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='r', s=0.1)
```

```python
plt.figure(figsize=(30, 10))
plt.subplot(131)
plt.scatter(points[:, 0], points[:, 1], c='r', s=0.1)
plt.subplot(132)
plt.scatter(points[:, 0], points[:, 2], c='r', s=0.1)
plt.subplot(133)
plt.scatter(points[:, 1], points[:, 2], c='r', s=0.1)
```

```python
boundary_points = cylinder.grid_points_on_boundary(50)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')
ax.scatter(
    boundary_points[:, 0],
    boundary_points[:, 1],
    boundary_points[:, 2],
    c='r',
    s=0.5)
```
