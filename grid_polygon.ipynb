```python
from geometry import Polygon
from geometry import Sampler
from matplotlib import pyplot as plt
import numpy as np

point = Polygon(False, *([0, 0], [0, 1], [1, 2], [2, 1], [2, 0]))

points = point.grid_points(100)
print(points.shape)
plt.figure(figsize=(10, 10))
plt.scatter(points[:, 0], points[:, 1], s=0.5, c='r')
```

```python
points = point.grid_points_on_boundary(100)
print(points.shape)
plt.figure(figsize=(10, 10))
plt.scatter(points[:, 0], points[:, 1], s=0.5, c='r')
```
