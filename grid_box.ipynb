```python
from Geometry import *
from matplotlib import pyplot as plt

box = Box([0, 0, 0], 4, 4, 10, q=[0, 0, 0])
points = box.grid_points(40)
print(points.shape)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')

ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=0.1, c='r')
```

```python
edge_points = box.grid_points_on_edge(40)
```

```python
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')
for key in edge_points:
    print(edge_points[key].shape)
    ax.scatter(edge_points[key][:, 0], edge_points[key][:, 1], edge_points[key][:, 2], s=0.1, c='r')
```
