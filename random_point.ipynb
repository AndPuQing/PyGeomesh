```python
from geometry import Point
from matplotlib import pyplot as plt

point = Point(False, [0, 0, 0], 2)
points = point.random_points(50000)
print(points.shape)

ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')

ax.scatter(
    points[:, 0],
    points[:, 1],
    points[:, 2],
    s=0.1,
    c='r',
)
```

```python
boundary_points = point.random_points_on_boundary(n=3000)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')

ax.scatter(
    boundary_points[:, 0],
    boundary_points[:, 1],
    boundary_points[:, 2],
    s=0.1,
    c='r',
)
```
