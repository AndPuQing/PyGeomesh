```python
from geometry import Point
from geometry import Polygon
from matplotlib import pyplot as plt
```

```python
point = Point(False, [0, 0], 1)
polygon = Polygon(False, [-2, 2], [2, 2], [2, -2], [-2, -2])

point_point = point.random_points(1000)
polygon_point = polygon.random_points(4000)

plt.figure(figsize=(10, 10))
plt.scatter(point_point[:, 0], point_point[:, 1], s=1)
plt.scatter(polygon_point[:, 0], polygon_point[:, 1], s=1)
```

```python
point_point_b = point.random_points_on_boundary(100)
polygon_point_b = polygon.random_points_on_boundary(400)

plt.scatter(point_point_b[:, 0], point_point_b[:, 1], s=1)
plt.scatter(polygon_point_b[:, 0], polygon_point_b[:, 1], s=1)
```

```python
polygon.difference(point)
```

```python
plt.scatter(polygon.points[:, 0], polygon.points[:, 1], s=1)
```
