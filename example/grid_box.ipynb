```python
from Geometry import *
from matplotlib import pyplot as plt
import numpy as np
box = Box([0, 0, 0], 4, 4, 10, q=[np.pi / 4, 0, 0])
points = box.grid_points(40)
print(points.shape)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')

ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=0.1, c='r')
```

```python
edge_points = box.grid_points_on_edge(20)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')
for key in edge_points:
    ax.scatter(
        edge_points[key][:, 0],
        edge_points[key][:, 1],
        edge_points[key][:, 2],
        s=0.1)
```

```python
surface_points = box.grid_points_on_surface(20)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')
for key in surface_points:
    ax.scatter(
        surface_points[key][:, 0],
        surface_points[key][:, 1],
        surface_points[key][:, 2],
        s=0.1)
```

```python
boundary_points = box.grid_points_on_boundary(20)
ax = plt.figure(figsize=(10, 10)).add_subplot(111, projection='3d')
for key in boundary_points:
    if key == 'num':
        continue
    ax.scatter(
        boundary_points[key][:, 0],
        boundary_points[key][:, 1],
        boundary_points[key][:, 2],
        s=0.1)
```
